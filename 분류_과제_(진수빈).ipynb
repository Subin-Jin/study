## Column 정보 ##

**기준년도** : 해당 정보의 기준년도를 제공함.  
**가입자일련번호** : 해당가입자에 부여한 일련번호(1 ~ 10,000)  
**시도코드** : 해당 수진자 거주지의 시도코드  
**성별코드** : 해당 정보 대상자의 성별을 제공함  
**연령대코드(5세단위)** : 기준년도에 수진자의 나이를 5세 단위로 그룹화(범주화)하여 구분한 코드  
**신장(5Cm단위)** : 검진자의 키(5CM 단위)  
**체중(5Kg 단위)** : 검진자의 몸무게(5KG 단위)  
**허리둘레** : 검진자의 허리둘레  
**시력(좌)** : 수검자의 좌측 눈의 시력  
**시력(우)** : 수검자의 우측 눈의 시력  
**청력(좌)** : 수검자의 좌측 귀의 청력  
**청력(우)** : 수검자의 우측 귀의 청력  
**수축기혈압** : 검진자의 최고 혈압으로 심장이 수축해서 강한 힘으로 혈액을 동맥에 보낼 때의 혈관 내압  
**이완기혈압** : 검진자의 최저 혈압으로 심장의 완기시의 혈압  
**식전혈당(공복혈당)** : 검진자 식사 전 혈당(혈액 100ml당 함유 되어 있는 포도당의 농도) 수치  
**총콜레스테롤** : 혈청 중의 에스텔형, 비에스테형(유리)콜레스테롤의 합  
**트리글리세라이드** : 단순지질 혹은 중성지질을 뜻함   
**HDL콜레스테롤** : HDL(고밀도 리포단백질)에 포함되는 콜레스테롤  
**LDL콜레스테롤** : LDL(저밀도 리포단백질)에 함유된 콜레스테롤  
**혈색소** : 혈액이나 혈구 속에 존재하는 색소단백으로 글로빈(globin)과 엠(heme)으로 구성되며 혈중의 산소운반체로서의 역할 수행  
**요단백** : 소변에 단백질이 섞여 나오는 것  
**혈청크레아티닌** : 크레아티닌은 크레아틴의 탈수물로 내인성 단백대사의 종말산물로서 신장에서 배설되고 그 증감은 음식물에 관계없이 근육의 발육과 운동에 관계함  
**(혈청지오티)AST** : 간 기능을 나타내는 혈액검사상의 수치, 간세포 이외에 심장, 신장, 뇌 , 근육 등에도 존재하는 효소로 이러한 세포들이 손상을 받는 경우 농도가 증가함  
**(혈청지오티)ALT** : 간 기능을 나타내는 혈액검사상의 수치, ALT는 주로 간세포 안에 존재하는 효소로, 간세포가 손상을 받는 경우 농도가 증가함  
**감마지티피** : 간 기능을 나타내는 혈액검사상의 수치, 간 내의 쓸개관(담관)에 존재하는 효소로 글루타민산을 외부에 펩티드나 아미노산 등으로 옮기는 작용을 함, 쓸개즙(담즙) 배설 장애, 간세포 장애 발생 시 혈중에 증가하게 됨  
**흡연상태** : 해당 수검자의 흡연 상태 여부  
**음주여부** : 해당 수검자의 음주 상태 여부  
**구강검진 수검여부** : 해당 검진자가 구강검진을 선택하여 검진하였는지 여부에 대한 항목  
**치아우식증유무** : 해당 수검자의 치아우식증 유무에 대한 항목   
**치석** : 해당 수검자의 치석 여부  
**데이터공개일자** : 데이터 작성 기준일자

### 컬럼 추가 및 제거 ###

```python
# '혈압분류' 컬럼 추가 : 정상혈압(수축기 혈압 120 미만 그리고 이완기 혈압 80 미만),
# 고혈압전단계 1기(수축기 120~129 또는 이완기 80~84), 2기(수축기 130~139 또는 이완기 85~89),
# 고혈압 1기(수축기 140~159 또는 이완기 90~99), 2기(수축기 160 이상 또는 이완기 100 이상),
# 출처 : 고혈압 가이드라인 - 임상진료지침정보센터 (https://www.guideline.or.kr › download_chronic)

temp = list(range(health_drop.shape[0]))
def new_col_creating(df_name):
    for i in range(health_drop.shape[0]):
        if health_drop.iloc[i]['수축기혈압'] >= 140 or health_drop.iloc[i]['이완기혈압'] >= 90:
            temp[i] = '고혈압'        
        elif 120 <= health_drop.iloc[i]['수축기혈압'] < 140 or 80 <= health_drop.iloc[i]['이완기혈압'] < 90:
            temp[i] = '고혈압전단계'
        elif health_drop.iloc[i]['수축기혈압'] < 120 and health_drop.iloc[i]['이완기혈압'] < 80:
            temp[i] = '정상혈압'
    return temp
 
health_drop['혈압분류']=new_col_creating(health_drop)
```

```python
# 비만자는 정상인보다 3배 이상 고혈압의 위험도가 높음. 관련 컬럼 추가
# 체질량지수(BMI) : 체중(kg) / 신장(meter)의 제곱
# WHO와 대한비만학회에서 과체중을 체질량지수 23kg/㎡ 이상, 비만을 체질량지수 25kg/㎡ 이상으로 정의하였음.

wght_meter2 = (health_drop['신장(5Cm단위)'] / 100)**2
health_drop['BMI'] = health_drop['체중(5Kg 단위)'] / wght_meter2
```

```python
# 랜덤포레스트, 로지스틱회귀, LightGBM 모델 사용하여 교차 검증 수행

from sklearn.model_selection import cross_val_score

from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.linear_model import LogisticRegression
from lightgbm import LGBMClassifier


rf_clf = RandomForestClassifier(n_estimators= 300 ,random_state=0)
lgst_clf = LogisticRegression(max_iter = 500, n_jobs = -1)
lgbm_clf = LGBMClassifier(n_estimators=400, learning_rate=0.05)

rf_scores = cross_val_score(rf_clf, X_train, y_train, cv=5, scoring='accuracy')
lgst_scores = cross_val_score(lgst_clf, X_train, y_train, cv=5, scoring='accuracy')
lgbm_scores = cross_val_score(lgbm_clf, X_train, y_train, cv=5, scoring='accuracy')

print('정확도')
print('랜덤포레스트: ', rf_scores)
print('로지스틱회귀: ', lgst_scores)
print('LightGBM: ', lgbm_scores)

print('정확도 평균')
print('랜덤포레스트: {:.3f}'.format(rf_scores.mean()))
print('로지스틱회귀: {:.3f}'.format(lgst_scores.mean()))
print('LightGBM: {:.3f}'.format(lgbm_scores.mean()))
```

```python
# 가장 정확도가 높은 랜덤포레스트 모델 사용. GridSearch로 하이퍼파라미터 튜닝 진행

params = {
    'max_depth': [4, 8, 16, 24, 32],
    'min_samples_leaf' : [1, 6, 12, 18],
    'min_samples_split' : [2, 8, 16, 24, 32]
}
# RandomForestClassifier 객체 생성 후 GridSearchCV 수행
rf_clf = RandomForestClassifier(n_estimators=100, random_state=0, n_jobs=-1)
grid_cv = GridSearchCV(rf_clf , param_grid=params , cv=2, n_jobs=-1 ) #n_jobs=-1로 설정하면 모든 cpu 코어를 다 씀
grid_cv.fit(X_train , y_train)

print('최적 하이퍼 파라미터:\n', grid_cv.best_params_)
print('최고 예측 정확도: {0:.4f}'.format(grid_cv.best_score_))
```
