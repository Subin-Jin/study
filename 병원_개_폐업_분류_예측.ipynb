```python
# 수익성비율 측정 : ROE, ROA
## 자기자본수익률(ROE) = 순이익 / 자기자본
## 자산수익률(ROA) = 세후영업이익 / 총자산

hosp_cln['ROE1'] = hosp_cln['profit1'] / hosp_cln['netAsset1']
hosp_cln['ROA1'] = (hosp_cln['revenue1'] - hosp_cln['salescost1'] - hosp_cln['sga1'] - hosp_cln['ctax1']) / (hosp_cln['liquidAsset1'] + hosp_cln['nonCAsset1'])
hosp_cln.head()
```

```python
sns.scatterplot(data = hosp_cln, x = 'ROE1', y = 'ROA1')
# ROE와 ROA 둘 다 수익성 지표로 어느 정도 양의 상관관계를 보여야 함. 두 지표가 크게 차이날 경우 차입금으로 인한 기업 위험도를 의미
```

```python
# 효율성비율 측정 : AT, ART
## 자산회전율(AT) = 매출액 / 연초 총자산
## 재고자산회전율(ART) = 매출원가 / 연초 재고자산
hosp_cln['AT1'] = hosp_cln['revenue1'] / (hosp_cln['liquidAsset2'] + hosp_cln['nonCAsset2'])
hosp_cln['ART1'] = hosp_cln['salescost1'] / hosp_cln['inventoryAsset2']
hosp_cln.head()
```

```python
# 부채비율 측정 : TD
## 총부채비율(TD) = 총부채 / 총자산
hosp_cln['TD1'] = hosp_cln['debt1'] / (hosp_cln['liquidAsset1'] + hosp_cln['nonCAsset1'])
hosp_cln.head()
```

```python
# 유동성비율 측정 : CR, QR
## 유동비율(CR) = 유동자산 / 유동부채
## 당좌비율(QR) = 당좌자산 / 유동부채
hosp_cln['CR1'] = hosp_cln['liquidAsset1'] / (hosp_cln['liquidLiabilities1'])
hosp_cln['QR1'] = hosp_cln['quickAsset1'] / (hosp_cln['liquidLiabilities1'])
hosp_cln.head()
```

```python
sns.scatterplot(data = hosp_cln, x = 'CR1', y = 'QR1')
# 당연하지만... 대놓고 양의 상관관계를 보임
```

### XGBoost ###

```python
# 예측 평가
from sklearn.metrics import confusion_matrix, accuracy_score
from sklearn.metrics import precision_score, recall_score
from sklearn.metrics import f1_score, roc_auc_score

def get_clf_eval(y_test, pred=None, pred_proba=None):
    confusion = confusion_matrix( y_test, pred)
    accuracy = accuracy_score(y_test , pred)
    precision = precision_score(y_test , pred)
    recall = recall_score(y_test , pred)
    f1 = f1_score(y_test,pred)
    # ROC-AUC 추가 
    roc_auc = roc_auc_score(y_test, pred_proba)
    print('오차 행렬')
    print(confusion)
    # ROC-AUC print 추가
    print('정확도: {0:.4f}, 정밀도: {1:.4f}, 재현율: {2:.4f},\
    F1: {3:.4f}, AUC:{4:.4f}'.format(accuracy, precision, recall, f1, roc_auc))
```

```python
xgb_wrapper = XGBClassifier(n_estimators=400, learning_rate=0.05, max_depth=3, eval_metric='logloss')
xgb_wrapper.fit(X_train, y_train, verbose=True)
w_preds = xgb_wrapper.predict(X_test)
w_pred_proba = xgb_wrapper.predict_proba(X_test)[:, 1]
```

```python
get_clf_eval(y_test , w_preds, w_pred_proba)
```

```python
# early stopping 30으로 설정
xgb_wrapper = XGBClassifier(n_estimators=400, learning_rate=0.05, max_depth=3)
evals = [(X_tr, y_tr), (X_val, y_val)]
xgb_wrapper.fit(X_tr, y_tr, early_stopping_rounds=30, eval_metric="logloss", 
                eval_set=evals, verbose=True) #early stopping 관련 내용은 fit 메소드 안에다가 파라미터로 넣어주면 됨

ws30_preds = xgb_wrapper.predict(X_test)
ws30_pred_proba = xgb_wrapper.predict_proba(X_test)[:, 1]
```

```python
get_clf_eval(y_test , ws30_preds, ws30_pred_proba)
```

```python
from xgboost import plot_importance
%matplotlib inline

fig, ax = plt.subplots(figsize=(10, 12))
plot_importance(xgb_wrapper, ax=ax)
```
